```python
from transformers import BartForConditionalGeneration, BartTokenizer

def read_pdf(file_path):
    with open(file_path, "rb") as file:
        text = ""
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()

    return text

def summarize_with_transformer(text):
    model_name = "facebook/bart-large-cnn"
    tokenizer = BartTokenizer.from_pretrained(model_name)
    model = BartForConditionalGeneration.from_pretrained(model_name)

    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=550, min_length=500, length_penalty=2.0, num_beams=4, early_stopping=True)

    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

try:
    pdf_path = "/content/Joel_Grus_Data_Science_from_Scratch_First_Principles_with_Python.pdf"
    pdf_text = read_pdf(pdf_path)
    summary = summarize_with_transformer(pdf_text)

    print("\nSummary:\n", summary)

except Exception as e:
    print(f"An error occurred: {e}")
```
